```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the dataframe
data= pd.read_csv('TSLA.csv', index_col= 'Date')
```

Performing the exploratory data analysis on the TESLA stock(2010-2020) 

```python
data.head()
```

```python
data.info()
```

```python
data['Volume'].plot(figsize=(25,7))
```

```python
#Ploting the stock prices of tsla form 2018 to 2020
df_18_20 = data.loc['02-01-2018':'03-02-2020']
#Plotting the original data and  the moving average for window size = 7
df_18_20['Close'].plot()
df_18_20['Close'].rolling(window=7).mean().plot()
```

```python

df_18_20['Open'].plot()
df_18_20['Open'].rolling(window=7).mean().plot()
```

Finding the relationship between adj close and volume

```python
df_18_20['Adj Close'].plot()
```

```python
#Plotting the scatter plot with adj close and volume
plt.scatter(df_18_20['Volume'], df_18_20['Adj Close'])
```

Calculating the corelation between Adj Close and Volume

```python
from scipy.stats import pearsonr,spearmanr
corr, _ = pearsonr(df_18_20['Adj Close'], df_18_20['Volume'])
print('Pearsons correlation: %.3f' % corr)
```

```python
corr, _ = spearmanr(df_18_20['Adj Close'], df_18_20['Volume'])
print('Spearmans correlation: %.3f' % corr)
```
